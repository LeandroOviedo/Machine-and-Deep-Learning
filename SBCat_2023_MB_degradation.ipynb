```python
import pandas as pd
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# load dataset
dataframe = pd.read_excel("MB_degradation__reaction_pathway.xlsx")
dataset = dataframe.values
df=dataframe
df
```

```python
df['Dye'] = zscore(df['[Dye]'])
df['Cat'] = zscore(df['[Catalyst]'])
df['pH'] = zscore(df['pH'])
df['time (min)'] = zscore(df['time (min)'])

# Convert to numpy - Classification
x_columns = df.columns.drop('Ref').drop('m/z').drop('18 w/ 44 always')
x = df[x_columns].values
y = df['m/z'].values

# Create train/test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Regression chart.
def chart_regression(pred, y, sort=True):
    t = pd.DataFrame({'pred': ypred, 'y': y_test.flatten()})
    if sort:
        t.sort_values(by=['y'], inplace=True)
    plt.plot(t['pred'].tolist(), 'bo--' , label='prediction')
    plt.plot(t['y'].tolist(), 'k^',label='expected')
    plt.ylabel('output')
    plt.legend()
    plt.show()

# Plot the chart
chart_regression(ypred.flatten(),y_train)
```
